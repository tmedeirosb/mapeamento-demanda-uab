```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo a função para plotar a contagem dos cursos com valores nas barras
def plot_course_count_with_values(column, title, dataframe):
    # Separando os cursos listados em cada resposta e contando a frequência de cada curso
    course_count = dataframe[column].str.split(';').explode().value_counts()
    
    # Criando um gráfico de barras para visualizar a contagem de cada curso
    plt.figure(figsize=(10,6))
    barplot = sns.barplot(y=course_count.index, x=course_count.values, palette='viridis')
    plt.title(title, fontsize=16)
    plt.xlabel('Contagem', fontsize=12)
    plt.ylabel('Curso', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    
    # Adicionando os valores em cada barra
    for p in barplot.patches:
        width = p.get_width()
        plt.text(width + 0.3, p.get_y() + p.get_height() / 2, f'{int(width)}', ha='center', va='center', fontsize=10)
        
    plt.show()

# Carregando o arquivo Excel
file_path = 'uab-ifrn-gestores-unico-anonimo.xlsx'  # Substitua pelo caminho correto do seu arquivo
xls = pd.ExcelFile(file_path)
df = xls.parse('Planilha1')  # Substitua pelo nome correto da sua planilha se necessário

# Nomes exatos das colunas
licenciatura_col = 'DENTRE OS CURSOS DE LICENCIATURA LISTADOS ABAIXO, SELECIONE QUAIS O SEU MUNICÍPIO TEM INTERESSE QUE SEJAM OFERTADOS? '
tecnologos_col = 'DENTRE OS CURSOS TECNÓLOGOS LISTADOS ABAIXO, SELECIONE QUAIS O SEU MUNICÍPIO TEM INTERESSE QUE SEJAM OFERTADOS? '
bacharelado_col = 'DENTRE OS CURSOS DE   BACHARELADO LISTADOS ABAIXO, SELECIONE QUAIS O SEU MUNICÍPIO TEM INTERESSE QUE SEJAM OFERTADOS? '
especializacao_col = 'DENTRE OS CURSOS DE   ESPECIALIZAÇÃO LISTADOS ABAIXO, SELECIONE QUAIS O SEU MUNICÍPIO TEM INTERESSE QUE SEJAM OFERTADOS? '

# Plotando os gráficos com valores nas barras
plot_course_count_with_values(licenciatura_col, 'Contagem dos Cursos de Licenciatura de Interesse dos Municípios', df)
plot_course_count_with_values(tecnologos_col, 'Contagem dos Cursos Tecnólogos de Interesse dos Municípios', df)
plot_course_count_with_values(bacharelado_col, 'Contagem dos Cursos de Bacharelado de Interesse dos Municípios', df)
plot_course_count_with_values(especializacao_col, 'Contagem dos Cursos de Especialização de Interesse dos Municípios', df)
```
